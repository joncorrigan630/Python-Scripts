### Defines a function that calculates the probability of flipping two heads in two simulated coin flips for n number of trials. 
### Visualizes how the law of large numbers influences the proportion of heads in the experiment. 
### Shows that as the number of trials increases, the experimental probability moves closer to the actual probability.    

```python
import numpy as np 
import matplotlib.pyplot as plt

def prob_coin_flip_experiment():
  # defining the two coins as lists
  coin1 = ['Heads', 'Tails']
  coin2 = ['Heads', 'Tails']
 
  # simulation of flipping both coins randomly
  coin1_result = np.random.choice(coin1)
  coin2_result = np.random.choice(coin2)
 
  # check if both coin flips are heads
  if coin1_result == 'Heads' and coin2_result == 'Heads':
    return 1
  else:
    return 0
 
# input number of trials in the experiment  
n = 100

# create empty lists for proportion of heads and number of flips
prop_heads = []
num_flips = []

# set counter to track the number of trials resulting in two heads
counter = 0
 
# iterate the experiment for n trials
for coin_flip in range(n):
  # if both coins are heads add 1 to the counter
  counter += prob_coin_flip_experiment()
  # track the proportion of two heads for each flip 
  prop_heads.append(counter/(coin_flip+1))
  # keep a list for number of flips
  num_flips.append(coin_flip+1)
 
font1 = {'family':'serif','color':'darkred','size':14}
font2 = {'family':'serif','color':'black','size':16}

# plot all flips and proportion of two heads
plt.plot(num_flips, prop_heads, marker='_', ms=1, mec='b', mfc='k', linewidth=2, label='Experimental Probability')
plt.title("Proportion of Two Heads in N Trials", fontdict = font2, loc='center') 
plt.xlabel('Coin Flips', fontdict = font1)
plt.ylabel('Proportion', fontdict = font1) 

plt.grid(color='k', linestyle='dotted', linewidth=0.25, axis='x') 
plt.grid(color='k', linestyle='dotted', linewidth=0.25, axis='y') 

plt.hlines(0.25, 0, n, colors='darkred', linestyle='dashed', linewidth=3, label='Actual Probability')
plt.legend()

plt.show() 
```
