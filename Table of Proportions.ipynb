```python
import pandas as pd 
import numpy as np
data = pd.read_csv("city_temperature.csv")
data.dropna(inplace=True) 
df = data.loc[0:, ["State", "City", "Month", "Day", "Year", "AvgTemperature"]] 
my_df = df.loc[df.State != "Additional Territories"] 
my_df 
```

```python
states_prop_1 = my_df['State'].value_counts()/len(my_df['State']) 
states_prop_1.head() 
```

```python
states_prop_2 = my_df['State'].value_counts(normalize=True)
states_prop_2.head() 
```

```python
# Create proportions - Excluding Null Responses
proportions = my_df['State'].value_counts(normalize=True)
print(proportions)
```

```python
# Create proportions - Include Null Responses
proportions_2 = my_df['State'].value_counts(normalize=True, dropna=False)
print(proportions_2) 
```
